<h3> 3. TRANSFORMAÇÃO DOS DADOS </h3>
<br>
<div style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">
    Como pode ser visto no resultado acima, tem-se exatamente <b>1.599.828</b> registros de vendas na base de dados. Contudo, os valores de venda de cada produto estão codificados em termos do zscore. Em estatística, o <a href=https://en.wikipedia.org/wiki/Standard_score>zscore</a> é uma medida que indica a distância em desvios padrão que um determinado valor está da média do conjunto. Em teoria, o domínio do zscore são todos os números reais, mas na prática aproximadamente 99,73% dos valores de um conjunto possuem zscore dentro do intervalo fechado [-3,3]. Desta forma, ir-se-á normalizar os zscores dos valores de venda dos produtos dentro do intervalo [0,1] usando a <a href=https://en.wikipedia.org/wiki/Feature_scaling> normalização min-max</a>. As equações a seguir provam que aplicar a normalização min-max no conjunto de zscores de uma variável é equivalente à se aplicar a mesma normalização no conjunto dos valores reais variável de interesse.
    
 Sejam <i>m, s, l, u e x</i> a média, o desvio padrão, o valor mínimo, o valor máximo e um valor qualquer do conjunto X respectivamente. Tem-se:
 </div>

$$ z = \frac{x - mean(X)}{sd(X)} \space (1); $$
<br>
<p style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">
Onde z representa o zscore de x. Sejam y e y', o valor x e do zscore de x normalizados respectivamente:
</p>
$$ y = \frac{x - l}{u - l} \space (2); $$
<br>
$$y' = \frac{z - min(Z)}{max(Z) - min(Z)} \space (3);  $$
<br>
<p style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">
Substituindo a equação (1) em (3), obtém-se:
<P>
$$y' = \frac{z - min(Z)}{max(Z) - min(Z)} = \frac{\frac{x-m}{s} - \frac{l-m}{s}}{\frac{u-m}{s} - \frac{l-m}{s}} = \frac{  \frac{x - \not{m} - (l- \not{m})}{\not{s}} }{ \frac{u- \not{m}-(l- \not{m})}{\not{s}}} = \frac{x - l}{u - l} = y$$
<br>
<p style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">
O resultado anterior mostra que normalizar o conjunto de zscores de uma variável é equivalente à se normalizar os valores da própria variável. Assim, é possivel obter uma ideia do futaramento que cada produto, produtor, categoria, entre outros gerou para a Hotmart sem se ter o valor real de cada venda.
</p>

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
sales['purchase_value_norm'] = min_max_scaler.fit_transform(sales[['purchase_value']].values)
sales[['product_id','purchase_value','purchase_value_norm']].sample(6)
```

<h4 style="text-indent:30px"> 4.1 MAIORES GERADORES DE VENDAS </h4>

<p style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">
    A seguir apresenta-se uma análise do total de vendas segmentada por produto.
    <li style="text-indent:30px"> <b> Segmentação por produto: </b> </li>
</p>

```python
# Cálculo do total de vendas por produto
sales_by_products = sales.groupby('product_id').count()[['purchase_id']].sort_values(by=['purchase_id'],
                                                                                       ascending=False)
sales_by_products.columns = ['total_by_product']
# Rankeamento dos produtos por quantidade de vendas
sales_by_products['rank_by_product'] = sales_by_products['total_by_product'].rank(ascending=False)
# Percentual do total de vendas por produto
sales_by_products['%by_product'] = sales['product_id'].value_counts(normalize=True) *100
# Ordenação  das colunas
sales_by_products = sales_by_products[['rank_by_product','total_by_product','%by_product']]
# -------------------------------------------------
sales_by_products.head(21)
```

<div style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">
    A tabela mostrada anteriormente reúne todos os produtos rankeados pelo número total de vendas. Os campos mostrados significam o seguinte: 
     <div style="margin-left: 40px">
        <li><b>product_id:</b> Identificação do produto na Hotmart;</li>
        <li><b>rank_by_product:</b> Ranking do produto por quantidade total de vendas ;</li>
        <li><b>total_by_product:</b> Quantidade total de vendas do produto; </li>
        <li><b>%by_product</b> Percentual de vendas do produtos em relação as vendas totais; </li>
      </div>
<div style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">
    Uma vez obtido o ranking de produtos mais vendidos, é possivel determinar qual é a revelância percentual no total de vendas de uma determinada quantidade de produtos do ranking.

```python
# Total de produtos distintos
len(sales_by_products)
```

```python
# Percentual cumulativo do total de vendas de cada produto do ranking
rank_cum = pd.DataFrame({'%':sales_by_products['%by_product'].cumsum()})

# Percentuais a serem analasidados
rankings = [2,5,20,60,120,240,450,800,1500,3000,5000]

# Output
percentuais = []
ranks = []
qtde_percentutal = []
i=0
for value in rankings:
    # Computa a quantidade de produtos dentro dentro de determinado percentual cumlativo
    text = "TOP {} produtos".format(str(value))
    percentuais.append(round(rank_cum.iloc[value,0]))
    ranks.append(text)
    # Computa a quantidade percentual de produtos 
    qtde_percentutal.append(round(100*len(rank_cum.loc[(rank_cum['%']<value)])/len(sales_by_products),2))

output = pd.DataFrame({'Rankings':qtde_products,'Qtde (%)':qtde_percentutal,
                      'Relevância (%)':percentuais})
output
```

<div style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">
    Os campos mostrados tabela acima significam o seguinte: 
     <div style="margin-left: 40px">
        <li><b>Rankings:</b> Rankings dos produtos mais vendidos;</li>
        <li><b>Qtde (%):</b> Quantidade percentual de produtos, em relação ao total de produtos distintos, dentro do ranking </li>
        <li><b>Relevância:</b> Relevância percentual do produto no total de vendas; </li>
    </div>  
    
<p style="text-align: justify; LINE-HEIGHT:30px; FONT-FAMILY:Arial;text-indent:30px">  Os resultados mostrados na tabela acima indicam, dentre outras coisas, que <b> os 57 produtos mais vendidos (cerca de 1,33% dos produtos disponíveis) são responsáveis por 50% de todas as vendas</b>. Um resultado bastante expressivo.
